```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(iris["sepal_length"],iris["sepal_width"])
```

```python
import seaborn as sns
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris)
```

```python
sns.lineplot(x= "petal_length",y="petal_width",data = iris)
```

```python
sns.distplot(iris["petal_width"])
```

```python
sns.barplot(x="sepal_length",y="sepal_width",data=iris)
```

```python
sns.boxplot(x="sepal_length",y="sepal_width",data=iris)
```

```python
sns.catplot(x="petal_length",y="petal_width",data=iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot=True)
```
